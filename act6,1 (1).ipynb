```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('NYPFellonies.csv')
df.head()
```

```python
sns.distplot(df.TOTAL)
plt.show()
```

```python
pd.pivot_table(df, 
               index='BOROUGH', 
               values='TOTAL', 
               aggfunc='sum'
              )
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
brgh = ['BX','BN','MN','QS','SI']
tot = [257,290,414,222,16]
ax.bar(brgh,tot)
plt.show()
```
